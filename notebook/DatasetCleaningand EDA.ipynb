```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset = pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv', header = 1)
```

```python
dataset.head()
```

```python
dataset.info()
```

## Data Cleaning

```python
## missing Values
dataset.isnull().sum()
# dataset.isnull().any(axis = 1)
```

```python
dataset[dataset.isnull().any(axis=1)]
```

```python
dataset.loc[:122, "Region"] = 0
dataset.loc[122:, "Region"] = 1
df = dataset
```

```python
df.info()
```

```python
df.head()
```

```python
df[['Region']] = df[['Region']].astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
## Removing the null values
df = df.dropna().reset_index(drop = True)
df.head()
```

```python
df.shape
```

```python
 df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
## Remove the 122nd row
df = df.drop(122).reset_index(drop = True)
```

```python
df.shape
```

```python
df.columns
```

```python
## Fix the spaces in the columns.str.strip()
df.columns = df.columns.str.strip()
df.columns
```

```python
df.info()
```

```python
df.columns
```

```python
## Chanages the required Columns as the integer type
df[['month','day', 'year', 'Temperature', 'RH', 'Ws']] = df[['month','day', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.info()
```

## Chnaging the other colums name is float datatype

```python
## Chnaging the other colums name is float datatype
objects = [features for features in df.columns if df[features].dtypes=='O']
```

```python
objects
```

```python
for i in objects :
    if i!='Classes':
          df[i] = df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
## Let save th ecleaned dataset
df.to_csv('Algerian_forest_fires_cleaned_dataset.csv', index = False)
```

```python
## Plot density plot for all the features
# plt.style.use('seaborn')
df_copy.hist(bins=50, figsize=(20, 15))
plt.show()
 
```

```python
## percentage of the pie chart
percentage = df_copy['Classes'].value_counts(normalize=True)*100
```

```python
## Ploting the pie chart
classlabels = ["Fire", "Not Fire"]
plt.figure(figsize= (12, 7))
plt.pie(percentage, labels=classlabels, autopct='%1.1f%%')
plt.title("Pie Chart of Classes")
plt.show
```

```python
sns.heatmap(df.select_dtypes(include=['number']).corr(), annot=True, cmap="coolwarm")
```

```python
plt.figure(figsize=(10, 8))  # Increase the figure size
sns.heatmap(
    df.select_dtypes(include=['number']).corr(), 
    annot=True,              # Show the correlation coefficients
    fmt=".2f",               # Format float to 2 decimal places
    cmap="coolwarm",         # Better color contrast
    cbar=True,               # Show color bar
    square=True,             # Make cells square
    linewidths=0.5,          # Add lines between boxes
    linecolor='gray',        # Color of the grid lines
    annot_kws={"size": 10}   # Font size of annotations
)
plt.title("Correlation Heatmap", fontsize=14)
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
